```python
# import libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler


# Create DataFrame
df = pd.DataFrame({'customer_id':[1,2,3,4,2,5],
                   'age': [25,np.nan, 35, np.nan, np.nan, 40],
                   'city': ['Mumbai;, ;Delhi', np.nan, 'Mumbai', 'Delhi', 'Mumbai', np.nan],
                   'gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
                   'annual_incom' : [40000, 80000, 60000, 120000, 80000, 90000]})

print("Original Data:\n",df)

#
#=====================
#====
# Task 1: Data Cleaning
#
#=====================
#====


# Fill missing values
df['age'].fillna(df['age'].median(), inplace=True)
df['city'].fillna(df['city'].mode()[0], inplace=True)

# Remove duplicate rows
df.drop_duplicates(inplace=True)

print("\nAfter Cleaning:\n", df)

#
#=====================
#====
# Task 2: Encoding
#
#=====================
#====

# One-Hot Encoding for city
df = pd.get_dummies(df,columns=['city'])

# Label Encoding for gender
le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])

print("\nAfter Encoding:\n",
df)

#
#=====================
# Task 3: Scaling
#
#=====================
#====

# Min-Max Scaling
minmax = MinMaxScaler()
df_minmax = df.copy()
df_minmax[['age', 'annual_incom']] = minmax.fit_transform(df_minmax[['age', 'annual_incom']])

print("\nMin-Max Scaled Data: \n", df_minmax)


# Standardization
standard = StandardScaler()
df_standard = df.copy()
df_standard[['age', 'annual_incom']] = standard.fit_transform(df_standard[['age', 'annual_incom']])

print("\nStandardized Data: \n", df_standard)
```
