```python
import requests
import time
import json
import os
from datetime import datetime

# Categories with keywords
categories = {
    "AI": ["ai", "machine learning", "deep learning"],
    "Startups": ["startup", "funding", "launch"],
    "Programming": ["python", "javascript", "code", "developer"],
    "Tech": ["tech", "software", "hardware"],
    "Others": []
}

BASE_URL = "https://hacker-news.firebaseio.com/v0"

# Function to assign category
def assign_category(title):
    title = title.lower()
    for category, keywords in categories.items():
        for word in keywords:
            if word in title:
                return category
    return "Others"

# Fetch top story IDs
try:
    response = requests.get(f"{BASE_URL}/topstories.json")
    top_ids = response.json()
except Exception as e:
    print("Error fetching top stories:", e)
    top_ids = []

all_stories = []

# Fetch stories (FIXED LOGIC)
for story_id in top_ids[:200]:  # take more to ensure 100+
    try:
        res = requests.get(f"{BASE_URL}/item/{story_id}.json")
        story = res.json()

        if not story or "title" not in story:
            continue

        assigned_category = assign_category(story["title"])

        data = {
            "post_id": story.get("id"),
            "title": story.get("title"),
            "category": assigned_category,
            "score": story.get("score", 0),
            "num_comments": story.get("descendants", 0),
            "author": story.get("by"),
            "collected_at": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        }

        all_stories.append(data)

        # Stop after enough stories
        if len(all_stories) >= 120:
            break

    except Exception as e:
        print(f"Error fetching story {story_id}: {e}")
        continue

# Sleep once (as required)
time.sleep(2)

# Create folder
os.makedirs("data", exist_ok=True)

# File name
filename = f"data/trends_{datetime.now().strftime('%Y%m%d')}.json"

# Save JSON
with open(filename, "w", encoding="utf-8") as f:
    json.dump(all_stories, f, indent=4)

# Final output
print(f"Collected {len(all_stories)} stories. Saved to {filename}")
```
