```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error

# Load dataset
df = pd.read_csv("/content/used_cars_messy.csv")

# Task 1
print(df.info())
print(df.describe())
print(df.shape)

# Task 2 - Cleaning

# Drop null target
df = df.dropna(subset=["selling_price"])

# Fill missing values
df = df.fillna(df.median(numeric_only=True))

# Clean brand column
df["brand"] = df["brand"].str.strip().str.lower()

# Clean mileage (example: "20 kmpl" → 20)
df["mileage"] = df["mileage"].str.extract('(\d+)').astype(float)

# Remove duplicates
df = df.drop_duplicates()

# Task 3 - Baseline MAE

mean_price = df["selling_price"].mean()

# Predict same value for all
y_true = df["selling_price"]
y_pred = [mean_price] * len(df)

mae = mean_absolute_error(y_true, y_pred)

print("Mean Price:", mean_price)
print("MAE:", mae)
```
